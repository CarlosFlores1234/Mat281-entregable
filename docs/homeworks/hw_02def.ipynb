```python
from google.colab import drive
drive.mount('/content/drive')


import os, glob, zipfile

BASE = "/content/drive/MyDrive/Tarea2_Aplica"
DATA_LP = f"{BASE}/data_learnplatform"

already = (
    os.path.isdir(DATA_LP) and
    os.path.exists(f"{DATA_LP}/districts_info.csv") and
    os.path.exists(f"{DATA_LP}/products_info.csv") and
    os.path.isdir(f"{DATA_LP}/engagement_data")
)
if not already:
    os.makedirs(DATA_LP, exist_ok=True)
    zips = [z for z in glob.glob(f"{BASE}/*.zip") if "learnplatform" in os.path.basename(z).lower()]
    if not zips:
        raise FileNotFoundError("No encontré el .zip de LearnPlatform en tu carpeta. Sube el ZIP o descomprime manualmente.")
    zip_path = max(zips, key=os.path.getsize)
    print("Descomprimiendo:", os.path.basename(zip_path))
    with zipfile.ZipFile(zip_path) as zf:
        zf.extractall(DATA_LP)

print("districts_info.csv:", os.path.exists(f"{DATA_LP}/districts_info.csv"))
print("products_info.csv :", os.path.exists(f"{DATA_LP}/products_info.csv"))
print("engagement_data/  :", os.path.isdir(f"{DATA_LP}/engagement_data"))



import pandas as pd, numpy as np, glob, re, os
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (9,4)

products  = pd.read_csv(f"{DATA_LP}/products_info.csv")
districts = pd.read_csv(f"{DATA_LP}/districts_info.csv")

def norm_cols(df):
    import re
    df = df.copy()
    mapping = {c: re.sub(r'[^a-z0-9]+', '_', c.lower()).strip('_') for c in df.columns}
    return df.rename(columns=mapping)

products  = norm_cols(products)
districts = norm_cols(districts)

sector_col = None
for c in products.columns:
    if 'sector' in c:
        sector_col = c
        break

pef_col = None
for c in products.columns:
    if 'primary' in c and 'essential' in c and 'function' in c:
        pef_col = c
        break

if sector_col is None:
    raise KeyError("No se encontró la columna de Sector (busqué algo que contenga 'sector'). "
                   "Imprime products.columns para ver el nombre exacto.")
if pef_col is None:
    raise KeyError("No se encontró la columna 'Primary Essential Function'. "
                   "Imprime products.columns para ver el nombre exacto.")

ax = (products[sector_col]
      .fillna('Unknown')
      .value_counts()
      .sort_values(ascending=False)) \
      .plot(kind='bar', color='#86c5da')
ax.set_title("Distribución de productos por Sector")
ax.set_xlabel("Sector"); ax.set_ylabel("Cantidad")
plt.xticks(rotation=45, ha='right'); plt.tight_layout()
plt.show()

func_major = (products[pef_col]
              .fillna('Unknown')
              .astype(str)
              .str.split(' - ').str[0])

ax = func_major.value_counts().head(12).plot(kind='bar', color='#ff9999')
ax.set_title("Top 12 categorías de 'Primary Essential Function'")
ax.set_xlabel("Función esencial"); ax.set_ylabel("Cantidad")
plt.xticks(rotation=45, ha='right'); plt.tight_layout()
plt.show()

districts = norm_cols(districts)
if 'district_id' not in districts.columns:
    print("Columnas districts:", districts.columns.tolist())
    raise KeyError("No se encontró 'district_id' en districts_info.csv")

state_col = None
for c in districts.columns:
    if c == 'state' or 'state' in c:
        state_col = c
        break
if state_col is None:
    print("Columnas districts:", districts.columns.tolist())
    raise KeyError("No se encontró la columna de estado en districts_info.csv")

state_map = districts.set_index('district_id')[state_col].to_dict()

rows = []
files = glob.glob(f"{DATA_LP}/engagement_data/*.csv")

sample_files = files[:200]
for fp in sample_files:
    m = re.search(r'(\d+)\.csv$', fp)
    if not m:
        continue
    did = int(m.group(1))
    st  = state_map.get(did)
    if pd.isna(st):
        continue
    try:
        df = pd.read_csv(fp, usecols=['engagement_index'])
    except Exception:
        continue
    en = df['engagement_index'].replace([np.inf, -np.inf], np.nan).dropna()
    if not en.empty:
        rows.append({'state': st, 'en': en.mean()})

eng = pd.DataFrame(rows)
state_mean = eng.groupby('state')['en'].mean().sort_values(ascending=False).head(20)

plt.figure(figsize=(8,6))
state_mean.plot(kind='barh', color='#69b3a2')
plt.gca().invert_yaxis()
plt.title("Engagement promedio por estado (muestra de distritos)")
plt.xlabel("Engagement promedio"); plt.ylabel("Estado")
plt.tight_layout(); plt.show()





```

```python
import pandas as pd, os

from google.colab import drive
drive.mount('/content/drive')

BASE = "/content/drive/MyDrive/Tarea2_Aplica"
DATA_TIT = os.path.join(BASE, "data_titanic")

train = pd.read_csv(os.path.join(DATA_TIT, "train.csv"))
test  = pd.read_csv(os.path.join(DATA_TIT, "test.csv"))


print(train.shape, test.shape)
train.head(3)


from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix

SEED = 2025
features = ["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
target   = "Survived"

X = train[features].copy()
y = train[target].astype(int)
X_test_kaggle = test[features].copy()

num_cols = ["Age","SibSp","Parch","Fare"]
cat_cols = ["Pclass","Sex","Embarked"]

pre = ColumnTransformer(
    transformers=[
        ("num", SimpleImputer(strategy="median"), num_cols),
        ("cat", Pipeline([
            ("imp", SimpleImputer(strategy="most_frequent")),
            ("oh", OneHotEncoder(handle_unknown="ignore"))
        ]), cat_cols),
    ],
    remainder="drop"
)

rf = Pipeline([
    ("pre", pre),
    ("clf", RandomForestClassifier(
        n_estimators=400, random_state=SEED, n_jobs=-1
    ))
])

Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.25, stratify=y, random_state=SEED)

rf.fit(Xtr, ytr)
pred = rf.predict(Xte)

acc = accuracy_score(yte, pred)
prec, rec, f1, _ = precision_recall_fscore_support(yte, pred, average="macro", zero_division=0)
print(f"RandomForest  Acc:{acc:.3f}  Prec:{prec:.3f}  Rec:{rec:.3f}  F1:{f1:.3f}")
print("Matriz de confusión:\n", confusion_matrix(yte, pred))


rf.fit(X, y)
pred_test = rf.predict(X_test_kaggle)

submission = pd.DataFrame({
    "PassengerId": test["PassengerId"],
    "Survived": pred_test.astype(int)
})
submission.to_csv("/content/titanic_submission.csv", index=False)
print("Listo: /content/titanic_submission.csv")
submission.head()



rf.fit(X, y)
pred_test = rf.predict(X_test_kaggle)

submission = pd.DataFrame({
    "PassengerId": test["PassengerId"],
    "Survived": pred_test.astype(int)
})
submission.to_csv("/content/titanic_submission.csv", index=False)
print("Listo: /content/titanic_submission.csv")
```

```python
import pandas as pd
import numpy as np

def add_features(df):
    out = df.copy()

    out["Title"] = out["Name"].str.extract(r",\s*([^\.]+)\.", expand=False).fillna("None")
    rare = ["Lady","Countess","Capt","Col","Don","Dr","Major","Rev","Sir","Jonkheer","Dona"]
    out["Title"] = out["Title"].replace({"Mlle":"Miss","Ms":"Miss","Mme":"Mrs"})
    out["Title"] = out["Title"].where(~out["Title"].isin(rare), "Rare")

    out["FamilySize"] = out["SibSp"] + out["Parch"] + 1
    out["IsAlone"] = (out["FamilySize"] == 1).astype(int)

    out["FarePerPerson"] = (out["Fare"] / out["FamilySize"]).replace([np.inf, -np.inf], np.nan)

    out["Age*Class"] = out["Age"] * out["Pclass"]

    out["CabinLetter"] = out["Cabin"].astype(str).str[0].replace("n", np.nan)

    return out

train_fe = add_features(train)
test_fe  = add_features(test)

features_fe = [
    "Pclass","Sex","Age","SibSp","Parch","Fare","Embarked",
    "Title","FamilySize","IsAlone","FarePerPerson","Age*Class","CabinLetter"
]
target = "Survived"

X = train_fe[features_fe]
y = train_fe[target].astype(int)
X_test = test_fe[features_fe]



from sklearn.model_selection import StratifiedKFold, GridSearchCV, train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_recall_fscore_support

SEED = 2025

num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X.columns if c not in num_cols]

pre = ColumnTransformer(
    transformers=[
        ("num", SimpleImputer(strategy="median"), num_cols),
        ("cat", Pipeline([
            ("imp", SimpleImputer(strategy="most_frequent")),
            ("oh", OneHotEncoder(handle_unknown="ignore"))
        ]), cat_cols),
    ],
    remainder="drop"
)

rf = RandomForestClassifier(random_state=SEED, n_jobs=-1)

pipe = Pipeline([("pre", pre), ("clf", rf)])

param_grid = {
    "clf__n_estimators": [300, 600],
    "clf__max_depth": [None, 6, 10],
    "clf__min_samples_split": [2, 5],
    "clf__min_samples_leaf": [1, 2],
    "clf__max_features": ["sqrt", "log2"]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

grid = GridSearchCV(
    pipe, param_grid, scoring="f1_macro", cv=cv, n_jobs=-1, verbose=0
)
grid.fit(X, y)

print("Mejor parámetro F1_macro (CV):", grid.best_score_)
print("Mejores hiperparámetros:", grid.best_params_)
best = grid.best_estimator_



Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.25, stratify=y, random_state=SEED)
best.fit(Xtr, ytr)
pred = best.predict(Xte)
acc = accuracy_score(yte, pred)
prec, rec, f1, _ = precision_recall_fscore_support(yte, pred, average="macro", zero_division=0)
print(f"[Mejor RF + FE] Acc:{acc:.3f}  Prec:{prec:.3f}  Rec:{rec:.3f}  F1:{f1:.3f}")


best.fit(X, y)
pred_test = best.predict(X_test).astype(int)

sub2 = pd.DataFrame({"PassengerId": test["PassengerId"], "Survived": pred_test})
sub2.to_csv("/content/titanic_submission_v2.csv", index=False)
print("Archivo listo: /content/titanic_submission_v2.csv")


!kaggle competitions submit -c titanic -f /content/titanic_submission_v2.csv -m "RF tuned + features"
```
